```python
# -------------------------------------------
# Comparison of degree bounds for Lagrange system
# -------------------------------------------

import math

# ---------- basic helpers ----------

def binom(n, k):
    return math.comb(n, k)

def factorial(n):
    return math.factorial(n)

# volume bound:
# vol_n(Q) <= d^n / n!
def volume_bound(n, d):
    return d**n / factorial(n)

# ---------- STRUCTURAL (UCT) BOUND ----------
# MV <= C(n,δ-i) * ν! * ( (vol_n(Q)/ρ^i)^{1/ν} + (1/ν!)^{1/ν} )^ν
# we use ρ = 1 as in the table

def structural_bound(n, p, i, d, rho=1):
    delta = n - p
    nu = n - i

    volQ = volume_bound(n, d)

    term = ( (volQ/(rho**i))**(1/nu) + (1/factorial(nu))**(1/nu) )**nu

    return binom(n, delta-i) * factorial(nu) * term


# ---------- MULTIHOMOGENEOUS BEZOUT ----------
# C(n-i, p-1) d^p (d-1)^{n-i-p+1}

def mh_bezout(n, p, i, d):
    return binom(n-i, p-1) * (d**p) * ((d-1)**(n-i-p+1))


# ---------- TOTAL DEGREE BEZOUT ----------
# d^{p+n+1}

def total_bezout(n, p, i, d):
    return d**(p+n+1)


# =========================================================
# TABLE 1 : Varying degree
# =========================================================

print("\n--- VARYING DEGREE (n=10, p=3, i=2) ---\n")

n = 10
p = 3
i = 2

for d in [2,3,5,10]:
    S = structural_bound(n,p,i,d)
    MH = mh_bezout(n,p,i,d)
    T = total_bezout(n,p,i,d)

    print(f"d={d:2d} | Structural={S:.3e} | MH={MH:.3e} | Total={T:.3e}")


# =========================================================
# TABLE 2 : Varying dimension
# =========================================================

print("\n--- VARYING DIMENSION (d=3, p=3, i=2) ---\n")

d = 3
p = 3
i = 2

for n in [6,8,10,12]:
    S = structural_bound(n,p,i,d)
    MH = mh_bezout(n,p,i,d)
    T = total_bezout(n,p,i,d)

    print(f"n={n:2d} | Structural={S:.3e} | MH={MH:.3e} | Total={T:.3e}")
```
